```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# 1. Define the system of first-order ODEs
def harmonic_oscillator(t, y, omega0, gamma):
    """
    y[0] is position (x)
    y[1] is velocity (v)
    Returns [dxdt, dvdt]
    """
    x = y[0]
    v = y[1]
    
    dxdt = v
    dvdt = -gamma * v - (omega0**2) * x
    
    return [dxdt, dvdt]

# 2. Set physical parameters
omega0 = 2.0  # Natural frequency
gamma = 0.5   # Damping coefficient (set to 0 for undamped simple harmonic motion)

# 3. Set simulation parameters
t_span = (0, 20)  # Start and end time (t0, tf)
t_eval = np.linspace(0, 20, 1000)  # Time points where we want the solution

# 4. Define initial conditions: [initial position, initial velocity]
y0 = [1.0, 0.0]  

# 5. Solve the initial value problem (IVP)
# Note: system parameters are passed via the 'args' tuple
sol = solve_ivp(
    harmonic_oscillator, 
    t_span, 
    y0, 
    t_eval=t_eval, 
    args=(omega0, gamma)
)

# 6. Extract the results
time = sol.t
position = sol.y[0]
velocity = sol.y[1]

# 7. Plot the results
plt.figure(figsize=(10, 5))
plt.plot(time, position, label='Position ($x$)', color='blue')
plt.plot(time, velocity, label='Velocity ($v$)', color='orange', linestyle='--')
plt.title('Damped Harmonic Oscillator Response')
plt.xlabel('Time ($t$)')
plt.ylabel('State Value')
plt.legend()
plt.grid(True)
plt.show()
```
